```julia
# 8a
digits = parse.(Int64, split(chop(read("8-input.txt", String)), " "))
# digits = parse.(Int64, split("2 3 0 3 10 11 12 1 1 0 1 99 2 1 1 2"))
```

```julia
function recur(i, meta_sum)  # Remaining digits, accumulator
    # This is always called at the start of a node
    childs = digits[i]
    n_meta = digits[i+1]
    println(i, ": node with ", childs, " children and ", n_meta, " metadata")
    i = i + 2
    for c in 1:childs  # Process any children
        (i, meta_sum) = recur(i, meta_sum)
        println(i, ": finished a child, sum at ", meta_sum)
    end
    # We are now at the start of a metadata section
    println(i, ": processing metas ", n_meta)
    meta_sum = meta_sum + sum(digits[i:i+n_meta-1])
    i = i + n_meta
    return (i, meta_sum)
end
```

```julia
recur(1, 0)
```

```julia
# 8b
function recur2(i)  # Remaining digits, accumulator
    # This is always called at the start of a node
    childs = digits[i]
    n_meta = digits[i+1]
    this_val = 0 
    println(i, ": node with ", childs, " children and ", n_meta, " metadata")
    i = i + 2
    child_vals = Int64[]
    for c in 1:childs  # Process any children
        (i, val) = recur2(i)
        push!(child_vals, val)
        println(i, ": finished a child, it had value ", val)
    end
    # We now determine the value for this node
    if childs == 0
        # Value is determine by the metadata sum
        this_val = sum(digits[i:i+n_meta-1])
    else 
        # The value is determined from looking up the value of the child
        # nodes
        for j in digits[i:i+n_meta-1]
            if j <= childs
                this_val = this_val + child_vals[j]
            end
        end
    end 
    i = i + n_meta
    return (i, this_val)
end
```

```julia
recur2(1)
```
